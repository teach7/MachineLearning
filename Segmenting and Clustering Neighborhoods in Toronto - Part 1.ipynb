# Segmenting and Clustering Neighborhoods in Toronto - Part 1

For this assignment, you will be required to explore and cluster the neighborhoods in Toronto.

Start by creating a new Notebook for this assignment.
Use the Notebook to build the code to scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe like the one shown below:

```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import csv
import json
import xml
import pandas as pd
```

Get the HTML of the Wikipedia page as a strin

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

Apply BeatifulSoup

```python
# make Soup using URL
soup = BeautifulSoup(url,'lxml')
# print(soup.prettify())
```

Access tags by treating them as attributes when using soup.find() First locate the table, then parce out each neighborhood using 'tr'

```python
neighborhood = soup.find('table', class_ = 'wikitable')
neighborhood_rows = neighborhood.find_all('tr')
```

Extract the columns 'Postcode' and 'Borough' and 'Neighbourhood' from TABLE

```python
info_list = []
for row in neighborhood_rows:
    info = row.text.split('\n')[1:-1] 
    info_list.append(info)

info_list
```

Create dataframe from table data

```python
df = pd.DataFrame(info_list[1:], columns=info_list[0])
df = df.rename(columns={df.columns[2]: "Neighborhood" })


df.head(12)
```

Ignore cells with a borough that is Not assigned.

```python
df = df[df.Borough != 'Not assigned']
df.reset_index(drop=True, inplace=True)

df.head(12)
```

Combine neighborhoods with the same postcode

```python
df = df.groupby(['Postcode', 'Borough'])['Neighborhood'].apply(','.join).reset_index()
df.columns = ['Postcode','Borough','Neighborhood']
df
```

Assigning Borough values to the Neignbourhood value is "Not assigned"

```python
df.loc[df['Neighborhood'] == 'Not assigned', 'Neighborhood'] = df['Borough']
```
